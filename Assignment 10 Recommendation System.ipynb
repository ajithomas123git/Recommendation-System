```python
#Feching ratings for Stardust
stardust_user_rating = data_new['Stardust']  
```

```python
stardust_user_rating
```

```python
#Finding the correlation with different books
similar_to_stardust= data_new.corrwith(stardust_user_rating) 
```

```python
similar_to_stardust.head(10)
```

```python
corr_star = pd.DataFrame(similar_to_stardust, columns=['Correlation'])
corr_star.dropna(inplace=True)
corr_star.head()
```

```python
corr_star[corr_star['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(5)  
# These are the books which we can recommend to UserID 1348
```
